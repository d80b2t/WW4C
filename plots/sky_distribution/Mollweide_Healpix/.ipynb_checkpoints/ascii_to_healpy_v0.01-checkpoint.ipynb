```python
import numpy as np
import healpy as hp
import math 
from astropy.io import ascii
import matplotlib.pyplot as plt
```

```python
path = '/cos_pc19a_npr/data/WISE/W4/'
#datafile = 'WISE_W4SNRge3_and_W4MPRO_lt_6.0_RADecls.dat'
datafile = 'WISE_W4_DetBitge8_RADeclsorted.tbl'
input_file = path+datafile
```

```python
data = ascii.read(input_file)  
```

```python
nSide = 512            ##  only the powers of 2 here                   
nPix = 12*(nSide)**2   ## Do not change this - always true
pixelMap = [0]*nPix
```

```python
print(nPix)
#print(data)
```

```python
ra = data['ra']
dec = data['dec']
pi = math.pi
```

```python
for row in data:
    l = row[0]
    b = row[1]

    theta = (-1.*b + 90.)/180.*pi
    phi = l/180.*pi
    pixelNum = hp.pixelfunc.ang2pix(nSide, theta, phi, nest=False)
    pixelMap[pixelNum] += 1

pixelMap = np.array(pixelMap)
```

```python
print((pixelMap))
print(type(pixelMap))
print(len(pixelMap))
```

```python
hp.mollview(pixelMap, coord='E')
```

```python
np.savetxt('WISE_W4_DetBitge8_pixelMap_nSide512.dat', pixelMap)
```
